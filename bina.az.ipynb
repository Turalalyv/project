```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
url='https://bina.az/'
response=requests.get(url)
bina=BeautifulSoup(response.text,'html.parser')
bina
```

```python
loc,price,date=[],[],[]
names=bina.findAll('div',{"class":"items-i"})
for i in names:
    loc.append(i.find('div',{'class':"location"}).text)
    price.append(i.find('div',{'class':"price"}).text)
    date.append(i.find('div',{'class':"city_when"}).text)
    
```

```python
all_bina=zip(loc,price,date)
all_bina
df=pd.DataFrame(all_bina,columns=['loc','price','date'])
```
