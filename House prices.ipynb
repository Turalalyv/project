```python
def clean_cordinates(col):
    col=str(col)
    col=col[:2]+'.'+col[2:]
    return float(col)
df['C1']=df['C1'].apply(clean_cordinates)
df['C2']=df['C2'].apply(clean_cordinates)
```

```python
df['Floor']=df['Mərtəbə'].apply(lambda x: int(str(x).split('/')[0]))
df['Mərtəbə']=df['Mərtəbə'].apply(lambda x: int(str(x).split('/')[0]))
df[['Qiymət', 'C1', 'C2', 'Kateqoriya', 'Mərtəbə', 'Floor', 'Otaq sayı','Kupça', 'İpoteka', 'Təmir','Sahə']]
```

```python
center=(40.366601,49.834965)
metroCordinates=[
    (40.3798499,49.8485742),
    (40.4004909,49.8506117),
    (40.4028298,49.8684909)]
import haversine as hs
def how_far(loc1,loc2):
    return(hs.haversine(loc1,loc2))
df['Center']=0
for i in range(len(df)):
    df['Center'][i]=1000*(how_far((df['C1'][i],df['C2'][i]),center))
```

```python
df['Metro_distance']=0
for i in range(len(df)):
    distance=[]
    for j in metroCordinates:
        distance.append(how_far((df['C1'][i],df['C2'][i]),j))
    
    df['Metro_distance'][i]=min(distance)*1000
```
