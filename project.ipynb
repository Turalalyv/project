```python
def model_validation(model,x_train_, x_test_, y_train_, y_test_):
    model.fit(x_train_,y_train_)
    predic=model.predict(x_test_)
    print('accuracy score is:', accuracy_score(y_test_,predic))
    print('precision score is:', precision_score(y_test_,predic))
    print('recall_score is:', recall_score(y_test_,predic))
    print( )
    return [model,precision_score(y_test_,predic)]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
x=df.drop('Survived',axis=1)
y=df['Survived']
xg=[]
for i in range(5):
    lr=LogisticRegression()
    lr2=XGBClassifier()
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
    print('Logistic:')
    model_validation(lr,x_train, x_test, y_train, y_test)
    print('Xgboost:')
    xg.append(model_validation(lr2,x_train, x_test, y_train, y_test))
    
```
